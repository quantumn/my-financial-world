## Approximation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x)+ 0.5*x
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
plt.plot(x, f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
# tag: sin_plot
# title: Example function plot
# size: 60
```

# Regression

#### Monomials as Basis Functions

```python
reg = np.polyfit(x, f(x), deg=1)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
# tag: sin_plot_reg_1
# title: Example function and linear regression
# size: 60
```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
# tag: sin_plot_reg_2
# title: Regression with monomials up to order 5
# size: 60
```
